```python
import pandas as pd
import os
```

```python
variables = ['DSND','DT00','DT32','EMNT','EMSD','EMXP']
month = ['January','february','March','April','May','June','July','August','September','October','November','December']
```

```python
for v in variables:
    for m in month:
        file = pd.read_csv('C://users/antoi/desktop/Gates_data/NOAA_cat_month/'+v+'/'+m+'_'+v+'.csv')
        shape_0 = file.shape[0]
        file = file[file['LAST_YEAR']>= 1970]
        shape_1 = file.shape[0]
        file = file.drop(file[((file['LAST_YEAR'] >= 2010) & (file['NB_VALUE'] < 4)) |
                      ((file['LAST_YEAR'] < 2010) & (file['LAST_YEAR'] >= 2000) & (file['NB_VALUE'] < 8)) |
                      (((file['LAST_YEAR'] < 2000) & (file['NB_VALUE'] < 10)))].index)
        shape_2 = file.shape[0]
        print(v+'--'+m+'--'+str(shape_0)+'--'+str(shape_1)+'--'+str(shape_2))
        print('_')
        file.to_csv('C://users/antoi/desktop/Gates_data/NOAA_cat_month/'+v+'/'+m+'_'+v+'_clean.csv')
        
```
