```python
import pandas as pd
import geopandas
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
%matplotlib inline
```

```python
gdf = geopandas.read_file('/tf/Team 6 - Project/Data/Shapefiles/06-shapefile/point.shp')
```

```python
gdf.columns
```

```python
gdf.shape
```

```python
gdf['ones'] = np.ones(gdf.shape[0])
```

```python
f = plt.figure(figsize=(40,20))
ax = f.add_subplot(111)
gdf.plot(ax = ax,column='TAVG',cmap = 'RdBu_r',markersize = 0.1)
plt.title('Temperature Average')
plt.savefig('/tf/Team 6 - Project/Results/06-TAVG.jpg')
plt.show()
```

```python
f = plt.figure(figsize=(40,20))
ax = f.add_subplot(111)
gdf.plot(ax = ax,column='ARH',cmap = 'BuGn',markersize = 0.1)
plt.title('Relative Humidity')
plt.savefig('/tf/Team 6 - Project/Results/06-ARH.jpg')
plt.show()
```

```python
f = plt.figure(figsize=(40,20))
ax = f.add_subplot(111)
gdf.plot(ax = ax,column='PRCP',cmap = 'Blues',markersize = 0.1)
plt.title('Precipitation')
plt.savefig('/tf/Team 6 - Project/Results/06-PRCP.jpg')
plt.show()
```
